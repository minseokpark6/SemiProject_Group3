```python
corr_matrix = temp.corr()
```

```python
plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, fontsize=8)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns, fontsize=8)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        text = "{:.2f}".format(corr_matrix.iloc[i, j])
        plt.annotate(text, xy=(j, i), ha='center', va='center')

plt.colorbar()

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# 예시 데이터 생성
X = temp.drop(columns=['출근대여량', '출근반납량', '총이용량_출근', '퇴근대여량', '퇴근반납량', '총이용량_퇴근', '출퇴근_총이용량'])
y = temp['출퇴근_총이용량']

# 다항 회귀 모델 설정
poly = PolynomialFeatures(degree=2)  # 추세선의 차수를 3으로 설정
X_poly = poly.fit_transform(X)

# 회귀 분석 결과 요약 구하는 코드
model = sm.OLS(y, X_poly)

result = model.fit()

print(result.summary())
```
