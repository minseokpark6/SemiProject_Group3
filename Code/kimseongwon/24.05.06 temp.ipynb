```python
temp = 출퇴근_이용량_통합.drop(['대여소_ID', '최종용도', '대여-반납량_출근', '대여-반납량_퇴근'], axis=1)
temp
```

```python
corr_matrix = temp.corr()
```

```python
plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, fontsize=8)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns, fontsize=8)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        text = "{:.2f}".format(corr_matrix.iloc[i, j])
        plt.annotate(text, xy=(j, i), ha='center', va='center')

plt.colorbar()

plt.show()
```

```python
import seaborn as sns
sns.pairplot(temp[['출근대여량', '출근반납량', '차이절대값_출근', '총이용량_출근', '퇴근대여량', '퇴근반납량', '차이절대값_퇴근', '총이용량_퇴근', '출퇴근_총이용량']])
plt.show()
```
