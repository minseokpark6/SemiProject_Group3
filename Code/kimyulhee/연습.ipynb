```python
import pandas as pd

# CSV 파일을 데이터프레임으로 읽기
data = pd.read_csv('C:/Users/evol0/OneDrive/바탕 화면/따릉이2/강서_따릉이_data_2023.csv')

# '기준_날짜' 열을 날짜 형식으로 변환 (형식이 'YYYY-MM-DD'인 경우)
data['기준_날짜'] = pd.to_datetime(data['기준_날짜'], format='%Y%m%d')

# 변환 후에 날짜 확인
print(data['기준_날짜'])
data.to_csv('C:/Users/evol0/OneDrive/바탕 화면/따릉이2/new_file3.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/evol0/OneDrive/바탕 화면/따릉이2/new_file3.csv', parse_dates=['기준_날짜'])

daily_usage = data.groupby('기준_날짜')['전체_건수'].sum()

plt.figure(figsize=(10, 6))
plt.plot(daily_usage.index, daily_usage.values, marker='o', linestyle='-')
plt.title('Daily Usage')
plt.xlabel('Date')
plt.ylabel('Usage Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/evol0/OneDrive/바탕 화면/따릉이2/new_file3.csv', parse_dates=['기준_날짜'])

daily_usage = data.groupby('기준_날짜')['전체_건수'].sum()

plt.figure(figsize=(10, 6))
plt.bar(daily_usage.index, daily_usage.values, color='skyblue')
plt.title('Daily Usage')
plt.xlabel('Date')
plt.ylabel('Usage Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
