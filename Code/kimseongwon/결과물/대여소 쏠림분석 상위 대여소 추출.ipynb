```python
강서_따릉이_데이터 = pd.read_csv('../../Data/따릉이/강서_따릉이_데이터_01.csv',encoding='utf-8')
```

```python
강서_따릉이_데이터
```

```python
출근 = 강서_따릉이_데이터[(강서_따릉이_데이터['기준_시간대'] >= 600) & (강서_따릉이_데이터['기준_시간대'] <= 1000)]
퇴근 = 강서_따릉이_데이터[(강서_따릉이_데이터['기준_시간대'] >= 1700) & (강서_따릉이_데이터['기준_시간대'] <= 2100)]

temp_df = 출근.groupby(['기준_날짜','시작_대여소_ID'])['전체_건수'].sum()
temp_df = pd.DataFrame(temp_df)
temp_df = temp_df.reset_index()
temp_df['기준_날짜'] = pd.to_datetime(temp_df['기준_날짜'])
temp_df.info()
```

```python
temp_df[:5]
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
# ADF 테스트는 시계열 데이터에서 비정상성을 나타내는 단위근의 존재 여부를 테스트하는 데 일반적으로 사용

def adf_test(data):
    result = adfuller(data) # adfuller 함수를 호출 >> ADF 테스트 진행
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')

adf_test(temp_df['전체_건수'])

# 결과 해석
# ADF의 귀무가설 '해당 시계열 데이터에 단위근이 존재한다 = 정상성이 없다'
# p-value: 7.857711085151761e-25 < 0.05이므로 귀무가설 기각
# 대립가설 '해당 데이터는 정상성을 가진 시계열 데이터이다' 채택

# ADF 통계: 테스트 통계 값은 -13.2761685272111입니다. 이 값은 시계열 데이터의 정상성을 평가하는 데 사용됩니다.
# 음수일수록 귀무가설(즉, 단위근의 존재)에 반하는 증거가 강해 정상성의 가능성이 더 높다는 것을 나타냅니다.

# p-값: 검정 통계량과 관련된 p-값은 대략 7.857711085151761e-25(또는 사실상 0)입니다.
# 이 p-값은 매우 작으며 일반적인 유의 수준인 0.05보다 훨씬 작습니다.
# 따라서 시계열에 단위근이 있다는 귀무가설을 자신있게 기각하고 데이터가 고정되어 있다는 결론을 내릴 수 있습니다.

# 임계 값: 1%, 5%, 10% 유의 수준의 임계 값입니다.
# 귀무 가설을 기각할지 여부를 결정하려면 ADF 통계를 이러한 임계 값과 비교해야 합니다.
# 이 경우 ADF 통계는 모든 임계값보다 훨씬 더 음수이므로 귀무 가설을 기각할 수 있습니다.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(copy['전체_건수'])
plot_pacf(copy['전체_건수'])
plt.show()

model = ARIMA(copy['전체_건수'], order=(1, 1, 0))  # p, d, q는 선택한 파라미터입니다.
model_fit = model.fit()

# 예측 결과 출력
print(model_fit.summary())

# t = model_fit.predict(start=0, end=len(copy)-1, typ='levels')

plt.plot(copy['전체_건수'].index, copy['전체_건수'].values, label='Actual')
plt.plot(copy['전체_건수'].index, model_fit.predict(start=0, end=len(copy)-1), label='Forecast')
plt.title('Actual vs. Predicted Bike Counts')
plt.xlabel('Date')
plt.ylabel('Bike Count')
plt.legend()
plt.grid(True)
plt.show()

forecast = model_fit.forecast(steps=1)
# print("다음 시간대 예측값:", forecast[0])
print(type(forecast))
print(forecast)
```
